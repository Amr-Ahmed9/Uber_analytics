```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the dataset
uber_data = pd.read_csv(r'F:\Users\Uber Analytics Dashboard\uber_ADB\uber-ride-analytics\data\raw\uber_rides_raw.csv', encoding='utf-8')
```

```python
# checking the shape of the dataset 
print("shape:",uber_data.shape)
```

```python
# loading name of attributes in dataset
print("\ncolmns:",uber_data.columns.tolist())
```

```python
print("\n columns data types:\n", uber_data.dtypes)
```

```python
# null_summary analysis
null_summary= pd.DataFrame({"missing count":uber_data.isnull().sum(),
                                   "missing percentage": uber_data.isnull().mean() * 100,
                                    "non-null count": uber_data.count(),
                                      "dtype": uber_data.dtypes}).sort_values(by="missing percentage", ascending=False)
print("\n null summary:\n", null_summary)
```

```python
# dublictaes analysis
print("\ntotal Duplicates:", uber_data.duplicated().sum())
print("dubicates Booking ID:", uber_data.duplicated(subset=['Booking ID']).sum())
print("dubicates Customer ID:", uber_data.duplicated(subset=['Customer ID']).sum())
print("Duplicate Booking IDs:", uber_data["Booking ID"].duplicated().sum())
```

```python
# catgorical attributes analysis
cat_col= uber_data.select_dtypes(include='object').columns
for feature_value in cat_col:
     # if you need to see all unique values in the column, you can uncomment the line below
     # print(f"\nUnique values in {col}:\n", uber_data[col].unique())
     #How many unique values exist, How often each value appears,Shows top 10 most common , you can use the line below
     print(f"\n{feature_value} ({uber_data[feature_value].nunique()} unique):\n", uber_data[feature_value].value_counts().head(10))
```

```python
# numerical attributes analysis
num_col= uber_data.select_dtypes(include=['int64', 'float64']).columns
for feature_value in num_col:
    print(f"\n{feature_value}:\n", uber_data[feature_value].describe())
```

```python
# outliers analysis for detected numerical attributes
num_cols =uber_data.select_dtypes(include=np.number).columns
for feature_value in num_cols:
    q1=uber_data[feature_value].quantile(0.25)
    q3=uber_data[feature_value].quantile(0.75)
    IQR=q3-q1
    outliers=uber_data[(uber_data[feature_value] < (q1 - 1.5 * IQR)) | (uber_data[feature_value] > (q3 + 1.5 * IQR))]
    if len(outliers) > 0:
       print(f"{feature_value}: {len(outliers)} outliers ({len(outliers)/len(uber_data)*100:.1f}%)")
```
